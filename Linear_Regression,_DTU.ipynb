## Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from sklearn.datasets import make_regression 
```

## Simple Linear Regression
- One feature

```python
x = np.linspace(0, 10, 100)
m = 30
c = 4
noise = np.random.randn(100)
y = m*x + c+ noise*10
yhat = m*x+c # prediction corresponding to x
```

```python
plt.scatter(x, y, label = "Data point")
plt.plot(x, yhat, c='red', label="Best fit line")
plt.xlabel("Feature, x")
plt.ylabel("Label, y")
plt.legend()
plt.show()
```

## Multiple Linear Regression
- More than one feature

```python
X, y, coef = make_regression(n_samples=100, n_features=2, noise=4, coef=True, bias=5)
print(X.shape, y.shape, coef)
```

```python
m1, m2 = coef
c = 5
x1, x2 = X[:,0], X[:,1] 
new_x1, new_x2 = np.linspace(X[:,0].min(), X[:,0].max(), 100), np.linspace(X[:,1].min(), X[:,1].max(), 50)
new_x1, new_x2 = np.meshgrid(new_x1, new_x2)
new_x1, new_x2 =  new_x1.reshape(-1, ), new_x2.reshape(-1, )
yhat = m1*new_x1 + m2*new_x2 + c
```

```python
plt.figure(figsize=(7,7))
ax = plt.axes(projection='3d')
ax.scatter(x1, x2, y) # actual data points
ax.scatter(new_x1, new_x2, yhat, c=yhat, s=5, alpha=0.5, cmap=plt.cm.coolwarm) # prediction surface 
ax.set_xlabel("Feature 1, x1")
ax.set_ylabel("Feature 1, x2")
ax.set_zlabel("Label, y")
plt.show()
```
